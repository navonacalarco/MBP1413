```python
#This script overlays all individual masks belonging to a target mask

import os
import numpy as np
from PIL import Image

# Path to the folder containing the mask images
folder_path = '/Users/home/Library/CloudStorage/OneDrive-UHN/archive/nucleiKaggle_2018/stage1_validation_masks/3c4c675825f7509877bc10497f498c9a2e3433bf922bd870914a2eb21a54fd26'

# Get a list of all mask image files in the folder
mask_files = [f for f in os.listdir(folder_path) if f.endswith('.png')]

# Initialize variables to store the dimensions of the masks
width = None
height = None
num_masks = len(mask_files)

# Iterate over each mask image
for i, mask_file in enumerate(mask_files):
    # Open the mask image
    mask_image = Image.open(os.path.join(folder_path, mask_file))

    # Convert the mask image to a NumPy array
    mask_array = np.array(mask_image)

    # Initialize width and height using the dimensions of the first mask
    if width is None or height is None:
        height, width = mask_array.shape
        # Initialize the overlay image with zeros
        overlay_image = np.zeros((height, width), dtype=np.uint8)

    # Add the mask to the overlay image
    overlay_image += mask_array

    print(f"Processed mask {i+1}/{num_masks}")

# Convert the overlay array back to an image
overlay_image = Image.fromarray(overlay_image)

# Save or display the resulting overlay image
overlay_image.show()  # Display the overlay image
overlay_image.save('/Users/home/Desktop/3c4c675825f7509877bc10497f498c9a2e3433bf922bd870914a2eb21a54fd26_overlay.png')  # Save the overlay image
```
